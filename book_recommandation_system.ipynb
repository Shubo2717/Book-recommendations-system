```python
add_ISBN = ratings.merge(books,on='ISBN')
```

```python
add_ISBN
```

```python
number_of_rating_df=add_ISBN.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_of_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
number_of_rating_df.head()
```

```python
average_rating_df=add_ISBN.groupby('Book-Title')['Book-Rating'].mean().reset_index()
average_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
average_rating_df.head()
```

```python
popular_df = number_of_rating_df.merge(average_rating_df,on='Book-Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df.head()
```

```python
popular_df=popular_df.merge(books,on='Book-Title')
```

```python
popular_df.head()
```

```python
popular_df.shape
```

```python
popular_df.duplicated('Book-Title').sum()
```

```python
print(popular_df.drop_duplicates('Book-Title',inplace=True))
```

```python
popular_df.shape
```

```python
popular_df.columns
```

```python
popular_df=popular_df[['Book-Title','Book-Author','Image-URL-M']]
```

```python
popular_df.head()
```

```python
x = add_ISBN.groupby('User-ID').count()['Book-Rating'] > 200
users_rating_more_than_200 = x[x].index
```

```python
filtered_rating = add_ISBN[add_ISBN['User-ID'].isin(users_rating_more_than_200 )]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
book_rating_more_than_49 = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(book_rating_more_than_49)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.head()
```

```python
pt.isnull().sum()
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.head()
```

```python
pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    recommended_books = []
    for i in similar_items:
        book_details = books[books['Book-Title'] == pt.index[i[0]]]
        recommended_books.append(book_details.iloc[0])
    recommended_df = pd.DataFrame(recommended_books)
    recommended_df.drop_duplicates('Book-Title',inplace=True)
    recommended_df=recommended_df[['Book-Title','Book-Author','Image-URL-M']]

    # recommended_json = recommended_df.to_json(orient='records')

    return recommended_df
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
recommend('1984')
```
